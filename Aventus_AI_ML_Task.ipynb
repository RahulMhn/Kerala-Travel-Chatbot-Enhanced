```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {'Hours_Studied': [2, 4, 5, 7, 8], 'Score': [21, 47, 53, 70, 85]}
df = pd.DataFrame(data)
X = df[['Hours_Studied']]
y = df['Score']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
hours_df = pd.DataFrame([[9]], columns=['Hours_Studied'])
predicted_score = model.predict(hours_df)
print(f"Predicted score for 9 hours: {predicted_score[0]:.2f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data') # Draws the dots
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.title('Hours Studied vs Student Score')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
